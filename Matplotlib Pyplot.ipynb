### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize graph
plt.figure(figsize=(5,3), dpi=100)

# Line 1

## Keyword Argument Notation
#plt.plot(x,y, label='2x', color='#eb4934', linewidth=1.5, linestyle='--', marker='.', markersize='4', markeredgecolor='black')

## Shorthand notation
## fnt = '[color][marker][linestyle]'
## Put it before label
plt.plot(x,y, 'bo-' ,label='2x')

# Line 2

## Select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

## Plot part of the graph as line
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')

## Plot remainder of the graph as a dashedline
plt.plot(x2[4:], x2[4:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('X and Y Graph', fontdict={'fontname': 'Arial Unicode MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis', fontdict={'fontname': 'Arial Unicode MS', 'fontsize': 12})
plt.ylabel('Y Axis', fontdict={'fontname': 'Arial Unicode MS', 'fontsize': 12})

# X and Y ticks (scale the graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

# Add a legeng
plt.legend()

# Save figure (dpi 300 has high resolution)
plt.savefig('xygraph.png', dpi=300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(5,3))

plt.show()
```
